```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
blank_img=np.zeros(shape=(720,720,3),dtype=np.int32)
```

```python
blank_img.shape
```

```python
plt.imshow(blank_img)
```

Writing text on image

```python
font=cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(blank_img,text='TEXT',org=(10,700),fontFace=font,fontScale=4,color=(255,255,255),thickness=3,lineType=cv2.LINE_4)
plt.imshow(blank_img)
```

Custom polygon

```python
vertices=np.array([[100,300],[200,200],[400,300],[200,400]],dtype=np.int32)
```

```python
vertices.shape
```

it's in (4,2), so we need to reshape into 3 dimensions

```python
pts=vertices.reshape((-1,1,2))
```

```python
pts.shape
```

```python
cv2.polylines(blank_img,[pts],isClosed=True,color=(0,0,255),thickness=5)
plt.imshow(blank_img)
```
